```python
# function to find dictionary with unique word along with there frequency. 
def word_freq_dict(document):
    word_freq = {}
    
    # Split the document into individual words
    words = document.split()
    
    # Loop through each word in the document
    for word in words:
        # Check if the word is already in the dictionary
        if word in word_freq:
            # If it is, increment the frequency
            word_freq[word] += 1
        else:
            # If it isn't, add it to the dictionary with a frequency of 1
            word_freq[word] = 1
    
    return word_freq
```

```python
# posting list
import os

def create_posting_list(path):
    posting_list = {}
    
    for filename in os.listdir(path):
            filepath = os.path.join(path, filename)
            with open(filepath, "r") as f:
                text = f.read()
                words = text.split()
                for word in words:
                    if word in posting_list:
                        if filename not in posting_list[word]:
                            posting_list[word].append(filename)
                    else:
                        posting_list[word] = [filename]
    
    return posting_list

```

```python
# #Binary Weighting scheme
import numpy as np

def binary(path):
  for i, file in enumerate(os.listdir(path)):
    with open(os.path.join(path, file), "r") as f:
        words = f.read().split()
        unique_words = set(words)
        word_freq = {}
        for word in words:
            if word in word_freq:
                word_freq[word] += 1
            else:
                word_freq[word] = 1
        for j, word in enumerate(words):
            tf = 1 if word_freq.get(word, 0) > 0 else 0  # use binary weighting scheme
            idf = idfVal[word]
            tf_idf = tf * idf
            dbinary.loc[file, word] = tf_idf
  return dbinary
  
```

```python
def count_frequency(filename, word_to_count):
    with open(filename, "r") as file:
        text = file.read()
        word_list = text.split()
        count = word_list.count(word_to_count)
    return count
```

```python
import pandas as pd
def rc(folder_path):
  # rowcount = pd.DataFrame(0, index=os.listdir(folder_path), columns=unique)
  for i, file in enumerate(os.listdir(folder_path)):
    with open(os.path.join(folder_path, file), "r") as f:
        words = f.read().split()
        unique_words = set(words)
        word_freq = {}
        for word in words:
            if word in word_freq:
                word_freq[word] += 1
            else:
                word_freq[word] = 1
        for word in word_freq:
            tf=count_frequency(file,word)
            idf=idfVal[word]
            tf_idf = tf*idf
            rowcount.loc[file,word]=tf_idf
  return rowcount
```

**Term Frequency**

```python
import pandas as pd
def termfrequency(folder_path):
    # dtf = pd.DataFrame(0, index=os.listdir(folder_path), columns=unique)
    for i, file in enumerate(os.listdir(folder_path)):
        with open(os.path.join(folder_path, file), "r") as f:
            words = f.read().split()
            unique_words = set(words)
            word_freq = {}
            max_freq = 1  # default value to avoid division by zero
            for word in words:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
                max_freq = max(max_freq, word_freq[word])
            for word in word_freq:
                tf = word_freq[word] / max_freq
                idf = idfVal[word]
                tf_idf = tf * idf
                dtf.loc[file, word] = tf_idf
    return dtf
```

**Log Normalization:**

```python
import pandas as pd
def lognorm(folder_path):
    # dln = pd.DataFrame(0, index=os.listdir(folder_path), columns=unique)
    for i, file in enumerate(os.listdir(folder_path)):
        with open(os.path.join(folder_path, file), "r") as f:
            words = f.read().split()
            unique_words = set(words)
            word_freq = {}
            for word in words:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
            for word in word_freq:
                tf = math.log(1 + word_freq[word])
                idf = idfVal[word]
                tf_idf = tf * idf
                dln.loc[file, word] = tf_idf
    return dln
```

**Double Normalization:**

```python
import pandas as pd
def dn(folder_path):
    # ddn = pd.DataFrame(0, index=os.listdir(folder_path), columns=unique)
    for i, file in enumerate(os.listdir(folder_path)):
        with open(os.path.join(folder_path, file), "r") as f:
            words = f.read().split()
            unique_words = set(words)
            word_freq = {}
            for word in words:
                if word in word_freq:
                    word_freq[word] += 1
                else:
                    word_freq[word] = 1
            max_freq = max(word_freq.values())
            for word in word_freq:
                tf = 0.5 + 0.5 * (word_freq[word] / max_freq)
                idf = idfVal[word]
                tf_idf = tf * idf
                ddn.loc[file, word] = tf_idf
    return ddn
```
